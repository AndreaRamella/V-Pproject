```python
#path = '/content/drive/MyDrive/Vision/Validation_Test/'
path = '/content/drive/MyDrive/Vision/Dataset48/'

folders_training = ['Arringatore', 'Atena', 'Atena armata', 'Demostene', 'Dioniso', 'Era Barberini', 'Ercole', 'Kouros_da_Tenea', 'Minerva_Tritonia', 'Poseidone', 'Zeus']
#folders_validation = ['Arringatore_valid', 'Atena_valid', 'Atena_armata_valid', 'Demostene_valid', 'Dioniso_valid', 'Era_barberini_valid', 'Ercole_valid', 'Kouros_da_tenea_valid', 'Minerva_tritonia_valid', 'Poseidone_valid', 'Zeus_valid']
#folders_test = ['Arringatore_test', 'Atena_test', 'Atena_armata_test', 'Demostene_test', 'Dioniso_test', 'Era_barberini_test', 'Ercole_test', 'Kouros_da_tenea_test', 'Minerva_tritonia_test', 'Poseidone_test', 'Zeus_test']
```

```python
from torchvision import transforms

transform = transforms.Compose([transforms.Resize(256), transforms.CenterCrop(256), transforms.ToTensor()])
 #transforms.CenterCrop(224),
 #transforms.Normalize( mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

change = transforms.ColorJitter(brightness=(0.5,1.5), contrast=(1), saturation=(0.5,1.5), hue=(-0.1,0.1))
```

```python
#appending the pics to the training data list
training_data = []
for lab in folders_training:
  print(lab)
  adress = path + lab
  for img in os.listdir(adress):
    img = Image.open(os.path.join(adress,img))
    img2 = change(img)
    img3 = change(img)
    img_t = transform(img)
    img_t2 = transform(img2)
    img_t3 = transform(img3)
    batch_t = torch.unsqueeze(img_t, 0)
    batch_t2 = torch.unsqueeze(img_t2, 0)
    batch_t3 = torch.unsqueeze(img_t3, 0)

    training_data.append(batch_t)
    training_data.append(batch_t2)
    training_data.append(batch_t3)
```

```python
#appending the pics to the training data list
valid_data = []
for lab in folders_validation:
  print(lab)
  adress = path + lab
  for img in os.listdir(adress):
    img = Image.open(os.path.join(adress,img))
    img_t = transform(img)
    batch_t = torch.unsqueeze(img_t, 0)

    valid_data.append(batch_t)
```
